```python
Patient_name=["A","B","C","D","E"]
Age=[20,68,30,49,90]
Condition=["Bad","Bad","Good","Good","Good"]
Oxygen_level=[80,68,95,90,98]
Available_beds=5
print("===Dataset Loaded===")
for i in range(len(Patient_name)):
    print("Patient_Name:",Patient_name[i],"Age:",Age[i],"Condition:",Condition[i],"Oxygen_Level:",Oxygen_level[i])
for i in range(len(Patient_name)):
    if Condition[i]=="Bad":
      if Age[i]>60  and Oxygen_level[i]<90:
        print("Allocate bed to patient:",Patient_name[i])
        Available_beds-=1
for i in range(len(Patient_name)):
      if Condition[i]=="Bad":
        if Available_beds>0:
          if Age[i]<60 and Oxygen_level[i]<90:
            print("Allocate bed to patient:",Patient_name[i])
            Available_beds-=1
for i in range(len(Patient_name)):
      if Condition[i]=="Bad":
        if Available_beds>0:
          if Age[i]>60 and Oxygen_level[i]>90:
            print("Allocate bed to patient:",Patient_name[i])
            Available_beds-=1


import matplotlib.pyplot as plt
plt.bar(Patient_name,Oxygen_level)
plt.xlabel("Patient Name")
plt.ylabel("Oxygen Level")
plt.title("Oxygen Level of Patients")
plt.show()
plt.bar(Patient_name,Age)
plt.xlabel("Patient Name")
plt.ylabel("Age")
plt.title("Age of patients")
plt.show()


```
