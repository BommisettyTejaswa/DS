```python

tank_capacity=1500
water_consumption=[50,75,100,50,70,100,200] # from day 1 to 7
avg_of_consumption=sum(water_consumption)/len(water_consumption)
print(avg_of_consumption)#avg consumption per day
avg_of_watercan_use=tank_capacity/len(water_consumption)
print(avg_of_watercan_use)
if avg_of_consumption>avg_of_watercan_use:
  print("Shortage of water")
else:
  print("No of shortage of water")
plt.bar(water_consumption,tank_capacity)
plt.title("Water consumption")
plt.xlabel("water_consumption per day")
plt.ylabel("tank_capacity")
plt.show()
```
