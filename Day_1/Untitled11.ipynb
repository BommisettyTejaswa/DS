```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
marks=[78,66,90,58,82,55,86,77,55,90]
hours=[7,6,9,3,8,5,7,4,8,9]
print("Mean:",np.mean(marks))
print("Median:",np.median(marks))
print("Standard Deviation:",np.std(marks))
print("Variance:",np.var(marks))
print("Range:",np.max(marks)-np.min(marks))
print("Correlation:",np.corrcoef(marks,hours))
plt.scatter(marks,hours)
plt.xlabel("Marks")
plt.ylabel("Hours")
plt.title("Marks vs Hours")
plt.show()



```
