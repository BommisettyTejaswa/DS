```python
import numpy as np
import matplotlib.pyplot as plt
Order_time=[1,1,2,1,3]
Delivery_time=[10,15,30,15,25]
Delivery_partner=["Swiggy","swiggy","zomato","Swiggy","Zomato"]
Customer_ratings=[5,4,3,4,3]
area=["North","East","West","East","South"]
temp=0
temp1=0
count=0
count1=0
count2=0
temp2=0
count3=0
temp3=0
print("===  Dataset Loaded ===")
for i in range(len(Order_time)):
    print("Order_time:",Order_time[i],"Delivery_time:",Delivery_time[i],"Delivery_partner:",Delivery_partner[i],"Customer_ratings:",Customer_ratings[i])
for i in range(len(Order_time)):
  if Delivery_partner[i]=="Swiggy":
    temp=temp+Delivery_time[i]
    count=count+1
avg_delivery_time=temp/count
print("Average Delivary time of swiggy:",avg_delivery_time)
for i in range(len(Order_time)):
  if Delivery_partner[i]=="Zomato":
    temp=temp+Delivery_time[i]
    count1=count1+1
avg_delivery_time1=temp/count1
print("Average Delivary time of zomato:",avg_delivery_time1)
most_delayed_partner=max(avg_delivery_time,avg_delivery_time1)
if most_delayed_partner==avg_delivery_time:
  print("Most delayed  is swiggy")
else:
  print("Most delayed  is zomato")
from collections import Counter
peak_order_hours = Counter(Order_time).most_common(1)[0][0]
print("Peak order hours:",peak_order_hours)
for i in range(len(Order_time)):
  if Delivery_partner[i]=="Swiggy":
    temp2=temp2+Customer_ratings[i]
    count2=count2+1
rating_ofswiggy=temp2/count2
print("Rating of swiggy:",rating_ofswiggy)
for i in range(len(Order_time)):
  if Delivery_partner[i]=="Zomato":
    temp3=temp3+Customer_ratings[i]
    count3=count3+1
rating_ofzomato=temp3/count3
print("Rating of zomato:",rating_ofzomato)
Top_rated_delivary_partner=max(rating_ofswiggy,rating_ofzomato)
if Top_rated_delivary_partner==rating_ofswiggy:
  print("Top rated delivary partner is swiggy")
else:
  print("Top rated delivary partner is zomato")
b=[]
for i in range(len(Order_time)):
  if Delivery_time[i]>20:
    b.append(area[i])
print("Most delayed areas are:")
for i in b:
  print(i)
a=[]
a.append(rating_ofswiggy)
a.append(rating_ofzomato)
Delivery_partner_labels = ["Swiggy", "Zomato"]
plt.bar(Delivery_partner_labels,a)
plt.xlabel("Delivary partner")
plt.ylabel("Customer ratings")
plt.title("Top ratings of Delivery partners ")
plt.show()
plt.scatter(Delivery_time,area)
plt.xlabel("Area")
plt.ylabel("Delivery time")
plt.title("Area vs Delivery time")
plt.show()
```
