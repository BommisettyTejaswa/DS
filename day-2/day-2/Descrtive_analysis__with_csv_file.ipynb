```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset from the CSV file
# df=data frame
df = pd.read_csv('/content/fitness_tracker.csv')
# Convert all columns to float (will error if there are text columns like 'Name' or 'Date')
df = df.astype(float)

print("====Fitness Center Dataset====")

# mean work out hours
average_workout = df['workout_hours'].mean()
print("\nAverage workout Hours=", average_workout)

# maximum calories
highest_calories = df['calories_burned'].max()
print("\nHighest Calories Burned=", highest_calories)

# minimum workout hours
lowest_workout = df['workout_hours'].min()
print("\nLowest Workout Hours=", lowest_workout)

# median calories
median_calories = df['calories_burned'].median()
print("\nMedian Colaries Burned=", median_calories)

# standard deviation
std_dev = df['calories_burned'].std()
print("\nStandard Deviation=", std_dev)

# most active member
most_active_member = df.loc[df['calories_burned'].idxmax(), 'member_names']
print("\nMost Active Member=", most_active_member)

# Plotting
plt.bar(df['member_names'], df['calories_burned'])
plt.title("Calories Burned by Members")
plt.xlabel("Member Names")
plt.ylabel("Calories Burned")
plt.show()
plt.barh(df['member_names'], df['calories_burned'])
plt.title("Calories Burned Over Time")
plt.xlabel("Calories Burned")
plt.ylabel("Member Names")
plt.show()
```
