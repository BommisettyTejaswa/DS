```r
data <- c(0,56,98,34,23,65,45,63,54,67,100)
print("Dataset")
print(data)
mean_value <- mean(data)
print(paste("Mean=",mean_value))
median_value <- median(data)
print(paste("Median=",median_value))
mode_function <- function(x) {
uniq_values <- unique(x)
uniq_values[which.max(tabulate(match(x,uniq_values)))]
}
mode_values <- mode_function(data)
print(paste("Mode=",mode_values))
max_value <- max(data)
print(paste("max=",max_value))
min_value <- min(data)
print(paste("min=",min_value))
variance_value <- var(data)
range_value <- max_value - min_value
print(paste("range=",range_value))
print(paste("variance=",variance_value))
standard_deviation_value <- sd(data)
print(paste("standard deviation=",standard_deviation_value))
summary(data)
barplot(data)
plot(data)
hist(data)
boxplot(data)





```
