```python
member_names=["Aman","Riya","Arjun","Rahul","Hemanth"]
ages=[22,24,25,26,24]
workout_hours=[2,1.5,2.5,1,3]
calories_burned=[500,350,600,300,750]
print("====Fitness Center Dataset====")
for i in range(len(member_names)):
  print(member_names[i],ages[i],workout_hours[i],calories_burned[i])
# mean work out hours
average_workout=sum(workout_hours)/len(workout_hours)
print("\nAverage workout Hours=",average_workout)
# maximum calories
highest_calories=max(calories_burned)
print("\nHighest Calories Burned=",highest_calories)
# miminumum ork out hours
lowest_workout=min(workout_hours)
print("\nLowest Workout Hours=",lowest_workout)
# median calories
import statistics
median_calories=statistics.median(calories_burned)
print("\nMedian Colaries Burned=",median_calories)
# standard deviation
std_dev=statistics.stdev(calories_burned)
print("\nStandard Deviation=",std_dev)
# most active member
index=calories_burned.index(highest_calories)
print("\nMost Active Member=",member_names[index])
import matplotlib.pyplot as plt
plt.bar(member_names,calories_burned)
plt.title("Calories Burned by Members")
plt.xlabel("Member Names")
plt.ylabel("Calories Burned")
plt.show()
```
